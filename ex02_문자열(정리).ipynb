# 문자열

```python
# 문자열 안에 작은 따옴표나 큰따옴표는 포함시키고 싶을 때 → 첫번째 방법
s = "she's gone"
s
```

```python
# he said that "she is gone"

# s1 변수와 s2 변수에 각각 두 방법을 활용하여 출력해보자!
s1 = 'he said that "she is gone"'
s2 = "he said that \"she is gone\""
print(s1)
print(s2)
```

### 문자열 인덱싱, 슬라이싱
- 인덱스(index) : 데이터가 생성될 때 자동으로 부여되는 순서 번호
- 인덱싱(indexing) : 무엇(인덱스)인가를 '가리킨다'는 의미
- 슬라이싱(slicing) : 무엇(인덱스 범위)인가를 '잘라낸다'는 의미
- Ctrl + a = 전체 열 선택

```python
# 문자열 인덱싱
msg = 'My name is BG'
print(msg[0])
print(msg[8])
```

```python
print(msg[-2])
print(msg[-1])
```

```python
# 문자열 슬라이싱 (※시작값은 해당 인덱스 지정, 끝값은 해당 인덱스 +1 지정)
# 'MY', 'name', 'is' 문자열 가져오기
print(msg[0:2])
print(msg[3:7])
print(msg[-5:-3])
```

```python
# 'My name' 문자열 가져오기
# 슬라이싱 문법 : [시작할값(인덱스):끝값(포함x)]
# 슬라이싱 문법2 : 비워두게(생략) 되면 → 처음부터 끝까지
print(msg[:7])
print(msg[3:])
```

```python
# 문자열 슬라이싱 실습
# Ctrl + 커서 : 다중입력 가능!

dy_info = "20200706Sunny"
year = "년도 : "
day = "월일 : "
weather = "날씨 : "
year+=dy_info[:4]
day +=dy_info[4:8]
weather +=dy_info[-5:]
print(year)
print(day)
print(weather)
```

```python
# 문자열을 이을때는 컴마 사용
print("년도 : ", dy_info[:4])
```

```python
# 문자열 포매팅(Formatting)
# 문자열 안의 특정한 값을 삽입해야 될 때 사용

day = 29
month = 12
s = '오늘은 %d월 %d일 입니다.'%(month,day)
s
```

### 문자열 포맷코드
- 문자열 내 값 삽입


```python
# .format() 함수를 사용한 포매팅!
# ※ " ,"→ 메소드 호출(함수 사용 시 앞에 "." 필수!), 함수 접근 시 필수 사용!

month = 11
day = 28
weather = "맑음"

s = "오늘은 {}월 {}일 {}입니다.".format(month,day,weather)
s
```

```python
# f 포매팅 을 사용한 포매팅!
# 안에서 연산 가능!
# 적용이 잘 되었을 때의 중괄호의 색은 검정색으로 변한다!(확인)

s = f"오늘은 {month}월 {day}일 {weather}입니다."
s
```

```python
# 변수 x에는 100을 대입, 변수 y에는 200을 대입 후 
# 변수 sum에는 두 변수의 합을 대입하고 포매팅을 이용하여 출력

x,y = 100,200
sum1 = x+y
print("%d과 %d의 합은 %d입니다."%(x,y,sum1))
print("{}과 {}의 합은 {}입니다.".format(x,y,sum1))
print(f"{x}과 {y}의 합은 {x+y}입니다.")
```
