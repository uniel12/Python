## matplotlib
- 그래프로 시각화 가능한 라이브러리
- 파이썬에서 데이터를 차트나 플롯으로 그려줌
- 차트의 종류는 산점도(scatter),막대(bar),원형(pie),꺽은 선,히스토그램 등 다양함

```python
import matplotlib.pyplot as plt
```

### 기본적인 그래프 그려보기

```python
# Line plot (선 그래프)
y = [2,12,6]

# plot(y요소)
# 데이터를 하나만 넣으면 y축으로 간다. 이때, x축은 인덱스 0부터 자동부여
plt.plot(y)

# show() -> 차트를 표현하는 print()
plt.show() # 사용권장
```

```python
# line chart(2개의 데이터)
x =[1,2,3,4,5]
y=[2,4,6,8,10]

plt.plot(x,y,ls=':',c='r',marker='o',lw=5,ms=15)

plt.show()
```

- 라인스타일 (Line Style) ls


- 라인컬러 (Line Color) c /color


- 마커 종류


- 옵션 종류


```python
# 실습 
import numpy as np
x =np.arange(8)
y=[5,2,5,6,8,7,5,7]
plt.plot(x,y,'r:D',mec='r',mfc='b',lw=3,ms=7)
plt.show()
```

- 결과


```python
x =np.arange(8)
y=[5,2,5,6,8,7,5,7]
# $특정문자$ -> 실제로 넣고 싶은 문자 기호
# color -> 16진수 코드로 가능
# 색상코드표 : https://www.rapidtables.org/ko/web/css/css-color.html
plt.plot(x,y,c='#1E90FF',ls=':',marker ='$❤$',mec='r',mfc='b',lw=3,ms=10)

# 그래프에 표시된 숫자의 범위지정
# 범위를 지정해서 그래프 확대나 축소 출력가능 - 차트를 한눈에 보고 싶을 때 사용
plt.xlim(-2,10)
plt.ylim(-5,12)

# 눈금을 지정
plt.xticks(np.arange(0,8,2))
plt.yticks(np.arange(2,9,2))

plt.show()
```

```python
x =np.arange(8)
y=[5,2,5,6,8,7,5,7]
# $특정문자$ -> 실제로 넣고 싶은 문자 기호
# color -> 16진수 코드로 가능
# 색상코드표 : https://www.rapidtables.org/ko/web/css/css-color.html
plt.plot(x,y,c='#1E90FF',ls=':',marker ='$❤$',mec='g',mfc='w',lw=1,ms=10)

# 눈금을 지정
plt.xticks(np.arange(0,8,2))
plt.yticks(np.arange(2,9,2))

# grid() : 격자무늬 생성
plt.grid()

plt.show()
```

## 여러 개의 꺾은 선 차트 하나의 공간에 그리기

```python
x = [2,3,4,5,6,7,8]
y1 = [5,7,3,9,1,8,6]
y2 = [9,5,7,2,3,7,1]

plt.xlabel('X')
plt.ylabel('Y',rotation = 0)

plt.plot(x,y1,c='r')
plt.plot(x,y2,c='b')
plt.show()
```

```python
# 한글이 깨진다... -> matplotlib 한글이 깨졌을 떄 사용!!!
from matplotlib import rc # 폰트를 설정하는 모듈
rc('font',family='Malgun Gothic')
# font : 폰트를 설정하겠다
# family : 글꼴 집합
```

```python
# 여러개의 차트 하나의 공간에 그려보기
x = [1,2,3,4]
y = [2,4,6,8]
z = [3,6,9,12]

plt.plot(x,label="x데이터")
plt.plot(y,label="y데이터")
plt.plot(z,label="z데이터")

# 범례 표시하기
plt.legend()

# chart 제목 부여
plt.title('제목',loc = 'right')
plt.show()
```

### 요일별 사고건수를 count해보자!
- 요일별로 사고건수를 확인하되, 요일 순서대로(월~일)출력

```python
# 요일별 발생한 사고건수를 카운트
data.columns
```

```python
y = data['요일'].value_counts()
y
```

```python
# 정렬을 하더라도 가나다 순을 되고, 요일별 정렬이 불가능함
y.sort_index()
```

```python
# 재정렬 하는 방법
y=y[['월','화','수','목','금','토','일']]
y
```

## Bar 차트

```python
x = y.index

plt.bar(x,y , color='lightblue') #bar 차트는 약자(c)로 못쓴다

plt.ylim(400,650)

plt.xlabel('요일')
plt.ylabel('사고건수', rotation =0)
plt.title('요일별 사고건수')

# text(x좌표,y좌표,넣고자 하는 값)
for i in range(len(y)) :
    plt.text(x[i],y[i],f'{y[i]}') # Bar , 꺾은선 가능

plt.show()
```

## '차대차' 사건 중 사상자 수가 많은 '발생지 시도' 를 알아보고 시각화
- 키워드 : 차대차, 사상자 수, 발생지 시도

```python
data.columns
```

```python
data['발생지시도']
```

```python
# 먼저 사고유형_대분류가 차대차인 행들만 출력
car = data[data['사고유형_대분류']=='차대차']
car
```

```python
# groupby(컬럼명).단일함수() 
# '발생지시도' 는 컬럼이 아니라 인덱스 별칭이 됨
chart_car = car[['발생지시도','사상자수']].groupby(['발생지시도']).sum()
chart_car
```

```python
# line 그리기
x = chart_car.index
y = chart_car['사상자수']
```

```python
# .figure((x,y)) -> chart 사이즈를 지정, 단 항상 차트 생성 위에!!!! 아래에 두면 적용X
plt.figure(figsize=(15,8))
plt.plot(x,y)
plt.show()
```

```python
# Bar 그리기 
x = chart_car.index
y = chart_car['사상자수']
```

```python
plt.figure(figsize=(15,8))
plt.bar(x,y)
plt.show()
```
