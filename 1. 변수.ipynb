#### 문자열 인덱싱과 슬라이싱
- 인덱싱 : 요소 하나를 가리키는 것-> 하나하나를 "가리킨다"
- 슬라이싱 : 요소들을 잘라오는 것 -> 범위를 지정해서, 어디부터 어디까지 "잘라낸다"
* 인덱스 번호는 앞에서 0부터, 뒤로세면 -1부터 시작
- 슬라이싱 : s[start:end] start값은 포함, end값은 미포함

```python
s="My name is SY"
print(s[0])
print(s[1])
print(s[-2])
print(s[-1])
print(s[0:2])
print(s[3:7])
print(s[8:10])
print(s[11:13])

print(s[0:2],'\n',s[3:7],'\n',s[8:10],'\n', s[11:13],sep="")
print(s[8:])
print(s[8:])
print(s[-5:])
print(s[:])
```
