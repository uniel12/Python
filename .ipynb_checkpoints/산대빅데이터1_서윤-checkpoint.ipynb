#### Question3
아래와 같이 리스트를 만들고 요소의 짝수/홀수를 구분하여 개수를 세는 프로그램을 작성하시오. 

```python
list1 = [1,50,27,19,37,33,5,3,48]

odd_cnt = 0
even_cnt = 0

for num in list1 :
    if num%2==1 :
        odd_cnt+=1
    else :
        even_cnt+=1

print(f'짝수 {even_cnt}개, 홀수 {odd_cnt}개')
```

#### Question4
다음 score_list에 들어있는 데이터를 이용해 아래와 같이 성적표를 만드는 프로그램을 작성하시오. (단, 학생의 숫자가 늘어나도 동작할 수 있도록 고려하시오.)

```python
import numpy as np
import pandas as pd

score_list = [['김미희',89,78,99],
              ['서대희',67,88,77],
              ['강예진',100,55,96]]

columns = ['이름','Python','Java','DB']

score =pd.DataFrame(score_list,columns = columns)
score= score.set_index('이름')
score
```

```python
score['평균']= score.loc[:,'Python':'DB'].mean(axis=1).astype(int)
score
```

```python
for i in score['평균']:
    if i>=90:
        score['학점']='A'
    elif i >= 80:
        score['학점']='B'
    elif i >= 70:
        score['학점']='C'
    elif i >=60 :
        score['학점']='D'
    elif i<60 :
        score['학점']='F'

score
```

```python
score.loc['평균']=score.loc[:'강예진','Python':'DB'].mean(axis=0).astype(int)
score
```

#### Question5
 X, Y1, Y2 리스트 자료를 활용하여 [실행결과]와 같은 그래프를 출력하는 프로그램을 작성하시오.  (단, 두 차트는 빨간색과 파란색을 통해 구별 하시오.)

```python
import matplotlib.pyplot as plt
x=[2,3,4,5,6,7,8]
y1=[5,7,3,9,1,8,6]
y2=[9,5,7,2,3,7,1]

plt.xlabel('X')
plt.ylabel('Y',rotation = 0)

plt.plot(x,y1,c='r',label='y1')
plt.plot(x,y2,c='b',label='y2')
plt.legend()
plt.grid()
plt.show()
```

#### Question6
Traffic_Accident_2017.csv 데이터를 이용하여 ‘차대 사람’인 사건 중 야간에 발생한 사건에 대해 아래 주어진 조건을 지켜 다음과 같이 시각화 하시오.

※ 조건 내용 반드시 확인할 것  
 조건 1. 사고유형_대분류가 ‘차대사람’ 이면서 발생시간대가 ‘야간’인 경우의 사고 건수 카운트  
 조건 2. y축의 범위는 0~200으로 설정할 것  
 조건 3. 차트 내부 글씨는 한글로 표시할 것  

```python
import numpy as np
import pandas as pd
import matplotlib.pyplot as plt
from matplotlib import rc
rc('font',family='Malgun Gothic')

data = pd.read_csv('./Traffic_Accident_2017.csv', encoding='euc-kr')
data
```

```python
data.info()
```

```python
data.columns
```

```python
data['사고유형_대분류']
```

```python
carPeo = data[data['사고유형_대분류']=='차대사람']
carPeo
```

```python
carPeo = carPeo[carPeo.loc[:,'주야']=='야간']
```

```python
y = carPeo['요일'].value_counts()
y = y[['월','화','수','목','금','토','일']]
y 
```

```python
x = y.index

plt.bar(x,y , color='#1976D2') #bar 차트는 약자(c)로 못쓴다

plt.ylim(0,200)

plt.xlabel('요일')
plt.ylabel('사고건수')
plt.title('요일별 야간 차대 사람 교통사고 건수')

    
plt.show()
```
