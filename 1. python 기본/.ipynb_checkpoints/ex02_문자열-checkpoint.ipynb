# 문자열

```python
# 문자열 안에 작은 따옴표나 큰따옴표는 포함시키고 싶을 때 → 첫번째 방법
s = "she's gone"
s
```

```python
# 문자열 안에 작은 따옴표나 큰따옴표는 포함시키고 싶을 때 → 두번째 방법
s1 = 'she\'s gone' 
s1
```

```python
# he said that "she is gone"

# s1 변수와 s2 변수에 각각 두 방법을 활용하여 출력해보자!
s1 = 'he said that "she is gone"'
s2 = "he said that \"she is gone\""
print(s1)
print(s2)
```

### 문자열 인덱싱, 슬라이싱
- 인덱스(index) : 데이터가 생성될 때 자동으로 부여되는 순서 번호
- 인덱싱(indexing) : 무엇(인덱스)인가를 '가리킨다'는 의미
- 슬라이싱(slicing) : 무엇(인덱스 범위)인가를 '잘라낸다'는 의미
- Ctrl + a = 전체 열 선택
* 인덱스 번호는 앞에서 0부터, 뒤로세면 -1부터 시작
- 슬라이싱 : s[start:end] start값은 포함, end값은 미포함

```python
s="My name is SY"
print(s[0])
print(s[1])
print(s[-2])
print(s[-1])
print(s[0:2])
print(s[3:7])
print(s[8:10])
print(s[11:13])

print(s[0:2],'\n',s[3:7],'\n',s[8:10],'\n', s[11:13],sep="")
print(s[8:])
print(s[8:])
print(s[-5:])
print(s[:])
```

```python
#문자열 실습 예제1
day="2020년 3월 3일의 날씨는 맑음입니다."
```

```python
print('날짜 : ',day[:11])
print('날씨 : ',day[-6:-4])
```

```python
# 문자열 슬라이싱 실습
# Ctrl + 커서 : 다중입력 가능!

dy_info = "20200706Sunny"
year = "년도 : "
day = "월일 : "
weather = "날씨 : "
year+=dy_info[:4]
day +=dy_info[4:8]
weather +=dy_info[-5:]
print(year)
print(day)
print(weather)
```

### 문자열 포맷 코드
- 문자열 내부에 값 삽입


#### 문자열 포매팅(Formatting)
- 문자열 안의 특정한 값을 바꿔야(삽입)할 경우가 있을 때 사용
    - %기호 포매팅
    - fomat 함수 포매팅
    - f 문자열 포매팅
    

```python
# ctrl + shifr +(-) 셀 나누기
# shift + m  셀병합
# 자동완성 tab키
```

```python
# 1. %기호 포매팅
day = 24
month =7

s= '오늘은 %d월 %d일 입니다.'%(month,day)
s
```

```python
# 2. fomat 함수 포매팅
day = 24
month =7

s= "오늘은 {}월 {}일 입니다.".format(month,day)
s
```

```python
# 3. f 문자열 포매팅
day = 24
month =7

s= f"오늘은 {month}월 {day}일 입니다."
s
```

```python
# 실습
x,y = 100,200
sum2 = x + y
print('%d와 %d의 합은 %d입니다.'%(x,y,sum2))
print('{}와 {}의 합은 {}입니다.'.format(x,y,sum2))
print(f'{x}와 {y}의 합은 {sum2}입니다.')
```
