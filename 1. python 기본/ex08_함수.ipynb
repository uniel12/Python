## 함수(function)
### 함수의 정의
    - 호출 될때 특정기능을 실행하는 <b>재사용 가능</b>한 코드 블록
### 함수의 목적
    - 반복적인 프로그램을 피할 수 있다.
    - 모듈화(코드모음)으로 인해 전체적인 코드의 가독성이 좋아진다.
    - 프로그램 문제가 발생하거나 기능의 변경이 필요할 때 유지 보수가 용이하다.

### 함수의 종류
1. 내장함수
2. 사용자정의 함수
3. 익명함수

#### 1. 내장함수
* print() : 출력하는 기능
* type() : 데이터 타입 반환하는 기능
* len() : 길이 확인하는 기능
* range() : 입력된 범위에서 일련의 숫자열을 생성하는 기능
* replace() : 문자열을 대체하는 기능
* sort() : 정렬하는 기능

```python
# 함수 정의
def number_sum(num1,num2):
    result = num1 + num2
    return result
```

```python
# 예제1
# 두 수를 입력 받아서 뺀 결과를 return하는 함수 정의하기
```

```python
def number_sub(num1,num2):
    if num1>num2:
        result= num1-num2
    else:
        result= num2-num1
    return result
```

```python
n1 = int(input('첫 번째 정수 입력 >> '))
n2 = int(input('두 번째 정수 입력 >>'))
print()
# 빼기 기능을 하는 함수 두 인자값 입력하여 호출
print(number_sub(n1,n2))
```

```python
# 예제2
# 두 수를 입력받아서 원하는 연산(+,-)을 수행하는 함수 정의하기
```

```python
def cal(num1,num2,op):
    """두 수를 입력받아서 원하는 연산을 수행하는 함수 """ # 독스트링
    if op =="+":
        result = num1+num2
    elif op =="-": 
            result= num1-num2 if num1>num2 else num2-num1
    else :
        result ="잘못 입력하셨습니다."
    return result
```

```python
n1 = int(input('첫 번째 정수 입력 >> '))
n2 = int(input('두 번째 정수 입력 >>'))
op = input("연산자 입력(+,-) >> ")
print()
result = cal(n1,n2,op)
print("결과 : {}".format(result))
```

### 함수의 특성
 1. 매개변수의 순서대로 인자가 들어감
 2. <b>기본값을 설정</b>해줄 수 있음

 4. <b>return의 결과값은 하나</b>
 5. <b>입력값</b>이나 <b>결과값</b>은 없을 수 있음
 * 결과값 : 변수에 담을 수 있거나 매개변수로 던져져야함

```python
def power_of_N(num,power=2):
    return num** power
```

```python
power_of_N(power=1,num=3) # 매개변수를 지정해서 호출 시 순서는 상관없음
```
