### 단축키 정리(커맨드 모드)
- 커맨드모드(파란색상태)
 - Enter : 에딧모드(초록색상태)로 전환
 - a : 기존 셀 위에 셀 삽입
 - b : 기존 셀 아래에 셀 삽입
 - m : 마크다운(주석)으로 전환
 - y : 마크다운에서 코드로 전환
 - dd: 해당 셀 삭제
 - x : 잘라내기
 - 3 : 마크다운 작성

### 단축키정리(에딧모드)
- 에딧모드(초록색상태)
 - Esc : 커맨드모드(파란색상태)로 전환
 - Ctrl+z : 되돌리기(※ 셀 안의 내용만 가능, 지운 셀 복구 불가!!)
 - 'tap'키로 불러오기 가능!

### 공통단축키
- Ctrl + Enter : 현재 셀 실행
- Alt + Enter : 셀 실행 후 아래에 셀 삽입
- Shift + Enter : 셀 실행 후 아래로 커서 이동
